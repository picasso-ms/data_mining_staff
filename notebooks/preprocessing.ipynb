```python
def min(mas):
    
    min= pd.Series()
    
    
    for i in wine_names[1:14]:
        m = np.copy(mas[i])
        min[i] = m[0]
        for j in range(1,177):
            if m[j]<min[i]:
                min[i] = m[j]
    return min
            
            
        
        
min(wine)
```

```python
def max(mas):
    
    max= pd.Series()
    
    
    for i in wine_names[1:14]:
        m = np.copy(mas[i])
        max[i] = m[0]
        for j in range(1,177):
            if m[j]>max[i]:
                max[i] = m[j]
    return max
            
            
        
        
max(wine)
```

```python
def getmed(mas):
    
    med= pd.Series()
    
    # size = mas.count()
    for i in wine_names[1:14]:
        size = mas[i].count()
        m = np.copy(mas[i])
        # m = mas[i].copy()
        # m.sort_values()
        m.sort()
        if size%2 == 0:
            xl = m[size//2]
            xr = m[size//2+1]

            # return (xl+xr)/2
            med[i]=(xl+xr)/2
            # print(size//2-1)
            # print(xl)
        else :
            s = size//2
            med[i]=s
            
            # return m[s]
    return med
pd.options.display.max_rows = 4000
getmed(wine)


```
