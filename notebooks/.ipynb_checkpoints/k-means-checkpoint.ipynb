```python
import pandas as pd
import numpy  as np
from numpy.random import random
from sklearn.utils import shuffle
from math import inf


names = ['sl','sw','pl','pw','class']
iris = pd.read_csv("../data/iris/iris.data",names=names)
iris.drop(columns='class',inplace=True)
iris = (iris - iris.min())/(iris.max()- iris.min())
iris = shuffle(iris)
irisV = iris.values
```

```python
c1 = pd.Series(random(4)).values
c2 = pd.Series(random(4)).values
c3 = pd.Series(random(4)).values
c = np.array( [c1,c2,c3]  )
centroids=[np.copy(c),0]
cond = [inf,inf,inf]

d1 = np.empty(0)
d2 = np.empty(0)
d3 = np.empty(0)
distance =np.empty((3,150)) 


while all(cnd > 0.005 for cnd in cond):
    distance =np.empty((3,150))
    centroids[0] = np.copy(c)
    for j in range(3):
        d = []
        for i in range(150):   
            sub = irisV[i]-c[j] 
            temp = np.dot( np.reshape(sub,(1,4)),np.reshape(sub,(4,1))  )
            d = np.append(d , np.sqrt(temp))
        distance[j] = d
    distance = np.reshape(distance,(150,3))
    iris = iris.reset_index(drop=True)
    species = []
    
    for i in range(150):
        species.append(distance[i].argmin())
    species
    iris['species'] = species
    for i in range(3):
        c[i] = iris[iris['species']==i].mean().drop('species')
    centroids[1] = np.copy(c)
    ll=[0,0,0]
    for i in range(3):
        s = centroids[1][i] - centroids[0][i]
        temp = np.dot(  np.reshape(s,(1,4)),np.reshape(s,(4,1)))
        cond[i] = temp[0][0]


print('Centroids: \n',c)
temp = [[],[],[]]
# temp = np.array( [np.empty(0),np.empty(0),np.empty(0)] )

for i in range(150):
    indx = distance[i].argmin()
    temp[indx].append(distance[i][indx])
    # temp[indx] = distance[i][indx]#np.append(temp[indx],distance[i][indx])
print(temp)


#TODO look in "metoda"
```
